```python
import pandas as pd
import numpy as np
```

```python
#reading data
data_matrix = pd.read_csv('mobile_data.csv')
#dropping id column
data_matrix = data_matrix.drop(['id'],axis=1)

print("features : ",len(data_matrix.columns)-1)
print("patterns : ",len(data_matrix))

low = data_matrix['price_range'].min()
high = data_matrix['price_range'].max()

print("Price range : ",low,"-",high)
```

```python
x = 0.1

# first = data_matrix.sample(n =20)
# print(first)
while(x<=0.9) : 
  x_data_matrix = data_matrix.sample(frac=x)
  y_data_matrix = data_matrix.drop(index = x_data_matrix.index)
  x = x+0.1
  print('******x data matrix ******\n',x_data_matrix)
  print('******y data matrix ******\n',y_data_matrix)
  print()
```

```python
print("Overall Mean ",data_matrix.stack().mean())
print("Overall Median ",data_matrix.stack().median())
print("Overall SD ",data_matrix.stack().std())
print("Overall Mode ",data_matrix.stack().mode())

print()

# print(data_matrix.mean(axis='rows'))
# print(data_matrix.mean(axis='columns'))


# print(data_matrix.median(axis='rows'))
# print(data_matrix.median(axis='columns'))

# print(data_matrix.mode(axis='rows'))
# print(data_matrix.mode(axis='columns'))

# print(data_matrix.std(axis='rows'))
# print(data_matrix.std(axis='columns'))

```
